```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

y0 = np.zeros(49, dtype=float)
y1 = np.zeros(49, dtype=float)
y2 = np.zeros(49, dtype=float)
for i in range(1, 8):
    input0 = f'out_0,0_{i}.txt'
    input1 = f'out_0,5_{i}.txt'
    input2 = f'out_1,0_{i}.txt'
    df = pd.read_csv(f'fruit-results/results-fruit-modif-include/{input0}', header=None)

    for j in range(0, 49):
        split = df[1][j].split(':')[0]

        if split == '4':
            y0[j] += (float(df[1][j].split(':')[1]))
        else:
            y0[j] += (float(df[2][j].split(':')[1]))

    df = pd.read_csv(f'fruit-results/results-fruit-modif-include/{input1}', header=None)

    for j in range(0, 49):
        split = df[1][j].split(':')[0]

        if split == '4':
            y1[j] += (float(df[1][j].split(':')[1]))
        else:
            y1[j] += (float(df[2][j].split(':')[1]))

    df = pd.read_csv(f'fruit-results/results-fruit-modif-include/{input2}', header=None)

    for j in range(0, 49):
        split = df[1][j].split(':')[0]

        if split == '4':
            y2[j] += (float(df[1][j].split(':')[1]))
        else:
            y2[j] += (float(df[2][j].split(':')[1]))

for i in range(0,49):
    y0[i] = y0[i] / 7
    y1[i] = y1[i] / 7
    y2[i] = y2[i] / 7

x = np.linspace(1,49,49, dtype=int)

fig, axs = plt.subplots(1, 3, figsize=(24, 6))

# plt.grid()
# plt.rcParams['figure.figsize'] = (12,6)
# plt.rcParams['figure.dpi'] = 200
# plt.rcParams.update({'font.size': 20})
# plt.xlabel('Selfish node mining power [%]')
# plt.ylabel('Selfish node profit [%]')
# plt.plot(x, y, 'r', linewidth=4.0, label='Selfish profit')
# plt.plot(x, x, 'g--', linewidth=4.0, label='Honest')
# plt.legend()

# plt.rcParams['figure.figsize'] = (14,8)
plt.rcParams['figure.dpi'] = 200
plt.rcParams.update({'font.size': 20})

fig.subplots_adjust(top=0.8)
fig.suptitle('Fruitchain include')

axs[0].grid()
axs[1].grid()
axs[2].grid()

axs[0].set_title('Gamma = 0.0')
axs[1].set_title('Gamma = 0.5')
axs[2].set_title('Gamma = 1.0')

axs[0].set_xlabel('Selfish node mining power [%]')
axs[1].set_xlabel('Selfish node mining power [%]')
axs[2].set_xlabel('Selfish node mining power [%]')

axs[0].set_ylabel('Selfish node profit [%]')
axs[1].set_ylabel('Selfish node profit [%]')
axs[2].set_ylabel('Selfish node profit [%]')

axs[0].plot(x, y0, 'r', linewidth=3.0, label='Selfish profit')
axs[0].plot(x, x, '--', color='gray', linewidth=2.0, label='Honest profit')
axs[0].legend()

axs[1].plot(x, y1, 'r', linewidth=3.0, label='Selfish profit')
axs[1].plot(x, x, '--', color='gray', linewidth=2.0, label='Honest profit')
axs[1].legend()

axs[2].plot(x, y2, 'r', linewidth=3.0, label='Selfish profit')
axs[2].plot(x, x, '--', color='gray', linewidth=2.0, label='Honest profit')
axs[2].legend()

plt.show()
fig.savefig('fruitchain_include.png')
```
